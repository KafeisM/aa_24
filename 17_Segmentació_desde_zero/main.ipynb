Una vega obtenim el conjunt de dades podem emprar-ho per fer dues tasques diferents: predicci贸 i entrenament. En ambdues tasques l'煤s de m猫triques 茅s fonamental, ja sigui per emprar-ho com a funci贸 de p猫rdua o b茅 com a mesura de qualitat.

A part de les m猫triques que ja coneixeu (IOU i Dice coefficient per segmentaci贸) existeix una funci贸 mpliament emprada per mesurar la qualitat d'aquest tipus de models: la **mAP**. Tot seguit teniu una explicaci贸 generada per ChatGPT i revisada per un hum:

La mAP (mean Average Precision) 茅s una m猫trica mpliament utilitzada per avaluar el rendiment en tasques de segmentaci贸, especialment en problemes on cal detectar i delimitar objectes dins d'una imatge. Aquesta m猫trica combina la precisi贸 i la recuperaci贸 per calcular l'efici猫ncia del model a l'hora de predir regions segmentades correctament.

En segmentaci贸, la mAP es calcula com la mitjana de les AP (Average Precision) per a diferents classes o nivells de superposici贸. Per a una sola classe, l'AP es deriva d'una corba de precisi贸-recall: es mesura la precisi贸 mitjana en diferents llindars de confian莽a o de superposici贸, com l'IoU (Intersection over Union). L'IoU quantifica la superposici贸 entre la mscara predita i la mscara real. Llindars comuns s贸n 0.5 (coincid猫ncia moderada) o valors m茅s alts per coincid猫ncies m茅s estrictes.

A la prctica, la mAP per segmentaci贸 sol incloure aquests passos:

- Calcul de l'IoU entre mscares predites i mscares reals.
- Associar prediccions correctes a les reals segons un llindar d'IoU establert.
- Generar la corba de precisi贸-recall.
- Obtenir l'AP per classe i, finalment, la mitjana de totes les classes per obtenir la mAP.


Una implementaci贸 de mAP extreta d'aquest [repositori](https://github.com/aladdinpersson/Machine-Learning-Collection/blob/master/ML/Pytorch/object_detection/metrics/mean_avg_precision.py)

````python
import torch
from collections import Counter

epsilon = 1e-9

def intersection_over_union(outputs, labels):
    outputs = outputs.squeeze(1)  # B x 1 x H x W => B x H x W

    intersection = (outputs & labels).float().sum((1, 2))
    union = (outputs | labels).float().sum((1, 2))

    iou = (intersection + epsilon) / (union + epsilon)

    return iou


def mean_average_precision(
        pred_boxes, true_boxes, iou_threshold=0.5, num_classes=20
):
    """
    Calculates mean average precision

    Parameters:
        pred_boxes (list): list of lists containing all bboxes with each bboxes
        specified as [train_idx, class_prediction, prob_score, x1, y1, x2, y2]
        true_boxes (list): Similar as pred_boxes except all the correct ones
        iou_threshold (float): threshold where predicted bboxes is correct
        num_classes (int): number of classes

    Returns:
        float: mAP value across all classes given a specific IoU threshold
    """

    # list storing all AP for respective classes
    average_precisions = []

    for c in range(num_classes):
        detections = []
        ground_truths = []

        # Go through all predictions and targets,
        # and only add the ones that belong to the
        # current class c
        for detection in pred_boxes:
            if detection[1] == c:
                detections.append(detection)

        for true_box in true_boxes:
            if true_box[1] == c:
                ground_truths.append(true_box)

        # find the amount of bboxes for each training example
        # Counter here finds how many ground truth bboxes we get
        # for each training example, so let's say img 0 has 3,
        # img 1 has 5 then we will obtain a dictionary with:
        # amount_bboxes = {0:3, 1:5}
        amount_bboxes = Counter([gt[0] for gt in ground_truths])

        # We then go through each key, val in this dictionary
        # and convert to the following (w.r.t same example):
        # ammount_bboxes = {0:torch.tensor[0,0,0], 1:torch.tensor[0,0,0,0,0]}
        for key, val in amount_bboxes.items():
            amount_bboxes[key] = torch.zeros(val)

        # sort by box probabilities which is index 2
        detections.sort(key=lambda x: x[2], reverse=True)
        TP = torch.zeros((len(detections)))
        FP = torch.zeros((len(detections)))
        total_true_bboxes = len(ground_truths)

        # If none exists for this class then we can safely skip
        if total_true_bboxes == 0:
            continue

        for detection_idx, detection in enumerate(detections):
            # Only take out the ground_truths that have the same
            # training idx as detection
            ground_truth_img = [
                bbox for bbox in ground_truths if bbox[0] == detection[0]
            ]

            best_iou = 0
            best_gt_idx = 0

            for idx, gt in enumerate(ground_truth_img):
                iou = intersection_over_union(
                    torch.tensor(detection[3:]),
                    torch.tensor(gt[3:]),
                )

                if iou > best_iou:
                    best_iou = iou
                    best_gt_idx = idx

            if best_iou > iou_threshold:
                # only detect ground truth detection once
                if amount_bboxes[detection[0]][best_gt_idx] == 0:
                    # true positive and add this bounding box to seen
                    TP[detection_idx] = 1
                    amount_bboxes[detection[0]][best_gt_idx] = 1
                else:
                    FP[detection_idx] = 1

            # if IOU is lower then the detection is a false positive
            else:
                FP[detection_idx] = 1

        TP_cumsum = torch.cumsum(TP, dim=0)
        FP_cumsum = torch.cumsum(FP, dim=0)
        recalls = TP_cumsum / (total_true_bboxes + epsilon)
        precisions = TP_cumsum / (TP_cumsum + FP_cumsum + epsilon)
        precisions = torch.cat((torch.tensor([1]), precisions))
        recalls = torch.cat((torch.tensor([0]), recalls))
        # torch.trapz for numerical integration
        average_precisions.append(torch.trapz(precisions, recalls))

    return sum(average_precisions) / len(average_precisions)
````
